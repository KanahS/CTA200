```python
import numpy as np
import rebound as rb
import reboundx as rx
import matplotlib.pyplot as plt
```

```python
def system_evolution(archive, extras):
    """ 
    Pulling data from the called upon simulation archive using REBOUND and REBOUNDx to collect evolution of
    the secondary binary in the simulation. 
    """ # UPDATE THIS DESCRIPTION
    sim = None
    sim = rx.SimulationArchive(archive, rebxfilename=extras)
    sim_time = []
    
    # SECONDARY BINARY
    b2_ecc = []
    b2_semi = []
    b2_period = []
    
    for snap in range(len(sim)):
        base = sim[snap][0]
        time = base.t
        
        # SECONDARY BINARY        
        b2_a = base.particles["Binary 2"].a
        b2_e = base.particles["Binary 2"].e
        b2_P = base.particles["Binary 2"].P
        
        b2_ecc.append(b2_e)
        b2_semi.append(b2_a)
        b2_period.append(b2_P)
        sim_time.append(time)
        
        #for key in base.particles.keys():
        #    print(key)
        
    #high_dir_cope = "/mnt/raid-cita/ksmith/cope_high-timestep/"
    
    #np.save(high_dir_cope+"sim_time", sim_time)
    #np.save(high_dir_cope+"b2_ecc", b2_ecc)
    #np.save(high_dir_cope+"b2_semi", b2_semi)
    #np.save(high_dir_cope+"b2_period", b2_period)
    
    return sim_time, b2_ecc, b2_semi, b2_period
```

```python
e_val = 0.473 # INITIAL PLANETARY SEMI-MAJOR AXIS UNDER EVALUATION
a_val = 5.000 # INITIAL BINARY ECCENTRICITY UNDER EVALUATION

print("e = ", e_val," a = ", a_val)
```

```python
###### TURN THIS INTO A FUNCTION THAT DOES EITHER BINARY OR PLANETS
high_sim_arch = "/mnt/raid-cita/ksmith/cope_high-timestep/sim_archive_Q10.0_eb0.473_ap5.000.bin"
high_xarch = "/mnt/raid-cita/ksmith/cope_high-timestep/xarchive_single_Qs10.bin"
```

```python
high_sys_evolution = system_evolution(high_sim_arch, high_xarch)
```

```python
fig, axs = plt.subplots(3, figsize=(10,10))

axs[0].plot(high_sys_evolution[0][0], high_sys_evolution[0][1])
axs[0].scatter(high_sys_evolution[0][0], high_sys_evolution[0][1])
axs[0].set_title(f"Binary Eccentricity w/ eb_initial = {e_val} and Q = {1e1}")
axs[0].set_ylabel("Eccentricity")

axs[1].plot(high_sys_evolution[0][0], high_sys_evolution[0][2])
axs[1].scatter(high_sys_evolution[0][0], high_sys_evolution[0][2])
axs[1].set_title(f"Binary Semi-Major Axis w/ ab_initial = 1 and Q = {1e1}")
axs[1].set_ylabel("Semi-Major Axis")

axs[2].plot(high_sys_evolution[0][0], high_sys_evolution[0][3])
axs[2].scatter(high_sys_evolution[0][0], high_sys_evolution[0][3])
axs[2].set_title(f"Binary Period w/ ab_initial = 1, eb_initial = {e_val}, and Q = {1e1}")
axs[2].set_ylabel("Period")

# SHARE X AXIS
# LABEL X AND Y AXES
plt.subplots_adjust(hspace=0.3)
```
